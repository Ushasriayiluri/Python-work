```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
data=pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
data['label']=breast_cancer.target
```

```python
x=data.drop(columns="label", axis=1)
y=data['label']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)
```

```python
print(x_train_std)
```

```python
pip install tensorflow
```

```python
#import tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up layer for neural network
model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20, activation="relu"),
                        keras.layers.Dense(30, activation="sigmoid"),
                        keras.layers.Dense(2, activation="sigmoid")
])
```

```python
#compiling the neural networks
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
#training neural network
history=model.fit(x_train_std, y_train, validation_split=0.1, epochs=10)
```

```python
#visualizing accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("model accuracy")
plt.ylabel('accuracy')
plt.xlabel("epochs")

plt.legend(['training_data', 'validation_data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("model loss")
plt.ylabel('loss')
plt.xlabel("epochs")

plt.legend(['training_data', 'validation_data'], loc='upper right')
```

```python
loss, accuracy=model.evaluate(x_test_std, y_test)
print(accuracy)
```
