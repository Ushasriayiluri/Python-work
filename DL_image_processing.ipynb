```python
#resizing the image using pillow library
from PIL import Image
```

```python
img=Image.open('puppy.jpg')
img_resized=img.resize((200,200))
```

```python
print(img)
```

```python
img_resized.save('puppy_resized.jpg')
```

```python
img=mpimg.imread('puppy_resized.jpg')
```
